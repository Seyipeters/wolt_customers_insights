# Customer Insights Analysis

This project analyzes customer data to extract insights about purchasing behavior, preferred restaurant types, and other key metrics. The analysis is performed using Python and Jupyter Notebook.

## Dataset
The dataset used for this analysis is dataset_for_datascience_assignment.csv. It contains customer-related information such as:
- Registration details
- Purchase counts
- Preferred restaurant types
- Total purchases in EUR
- Delivery and takeaway preferences
- Purchase times and frequencies

## Key Insights

### 1. Total Purchases Over Time
A line chart was created to visualize the trend of total purchases over time. This helps identify patterns in customer spending behavior.

### 2. Preferred Restaurant Types
A bar chart was used to analyze the total purchases (in EUR) by preferred restaurant types. This provides insights into customer preferences.

### 3. Most Engaged Users by Country
A bar chart was plotted to show the number of users by country. Finland was identified as having the most engaged users.

### 4. Purchase Frequency Segmentation
Customers were segmented based on their average days between purchases:
- Frequent: 0–7 days
- Moderate: 8–30 days
- Infrequent: 31+ days

The average and total spending for each segment were calculated and visualized.

### 5. Spending by Device
The average spending by preferred device (e.g., Android, iOS, Web) was analyzed and visualized using a bar chart.

### 6. Delivery vs Takeaway Preferences
A pie chart was created to compare the proportion of purchases made via delivery versus takeaway.

### 7. Purchases by Meal Type
The total purchases for different meal types (e.g., breakfast, lunch, dinner) were visualized using a bar chart.

### 8. Purchases by Hour of the Day
A bar chart was plotted to show the number of purchases made during different hours of the day.

## Tools and Libraries
The following Python libraries were used:
- pandas for EDA
- matplotlib and seaborn for data visualization


### Results and Recommendations

Based on the insights from the analysis and charts, here are some actionable recommendations:

#### 1. Total Purchases Over Time
- **Insight**: The trend of total purchases over time shows patterns in customer spending behavior.
- **Recommendation**:
    - Identify peak periods and align marketing campaigns or promotions with these times to maximize revenue.
    - Investigate dips in purchases to understand potential causes and address them.

#### 2. Preferred Restaurant Types
- **Insight**: Certain restaurant types generate higher total purchases.
- **Recommendation**:
    - Focus on promoting the most preferred restaurant types to attract more customers.
    - Collaborate with restaurants in the most popular categories to offer exclusive deals.

#### 3. Most Engaged Users by Country
- **Insight**: Finland has the most engaged users.
- **Recommendation**:
    - Invest in targeted marketing campaigns in Finland to further engage this audience.
    - Explore opportunities to replicate successful strategies in other countries with lower engagement.

#### 4. Purchase Frequency Segmentation
- **Insight**: Customers are segmented into frequent, moderate, and infrequent purchasers.
- **Recommendation**:
    - Offer loyalty programs or discounts to frequent purchasers to retain them.
    - Create targeted campaigns to convert moderate and infrequent purchasers into frequent buyers.

#### 5. Spending by Device
- **Insight**: Spending varies by preferred device but the IOS users has more spending.
- **Recommendation**:
    - Optimize the user experience on the most popular devices to encourage higher spending.
    - Develop device-specific promotions to engage users across all platforms.

#### 6. Delivery vs Takeaway Preferences
- **Insight**: Delivery purchases are more common than takeaways.
- **Recommendation**:
    - Enhance delivery services by improving speed, reliability, and customer satisfaction.
    - Offer incentives for takeaway orders to balance preferences and reduce delivery costs.

#### 7. Purchases by Meal Type
- **Insight**: Certain meal types (e.g., lunch, dinner) generate more purchases.
- **Recommendation**:
    - Focus marketing efforts on the most popular meal times to maximize sales.
    - Introduce meal-specific promotions (e.g., discounts on breakfast orders) to boost less popular categories.

#### 8. Purchases by Hour of the Day
- **Insight**: Purchases are concentrated during specific hours of the day.
- **Recommendation**:
    - Schedule promotions or discounts during off-peak hours to encourage more purchases.
    - Ensure adequate staffing and resources during peak hours to handle demand efficiently.

#### General Recommendations
- **Data-Driven Decisions**: We can use the above insights and recommendations to guide business strategies, such as pricing, promotions, and partnerships.
- **Customer Retention**: Implement loyalty programs and personalized offers to retain high-value customers.
- **Expand Market Reach**: Focus on underperforming segments or regions to grow the customer base.

These recommendations can help improve customer engagement, optimize operations, and increase revenue.

```python
#Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates

# Import seaborn
import seaborn as sns
# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
#Change the date time datatype
customers['REGISTRATION_DATE'] = pd.to_datetime(customers['REGISTRATION_DATE'])
customers['FIRST_PURCHASE_DAY'] = pd.to_datetime(customers['FIRST_PURCHASE_DAY'])
customers['LAST_PURCHASE_DAY'] = pd.to_datetime(customers['LAST_PURCHASE_DAY'])

#Check the data types of the columns again
customers.info()
```

```python
#Change the columns having object data types to string
for col in customers.columns:
    if customers[col].dtype == 'object':
        customers[col] = customers[col].astype(str)
        
#change the columns having object data types to string
customers['REGISTRATION_COUNTRY'] = customers['REGISTRATION_COUNTRY'].astype(str)
customers['PREFERRED_DEVICE'] = customers['PREFERRED_DEVICE'].astype(str)
customers['PREFERRED_RESTAURANT_TYPES'] = customers['PREFERRED_RESTAURANT_TYPES'].astype(str)


        
```

```python
customers.info()
```

```python
customers.isnull().sum()
```

```python
#Check the percentage of missing values in each column
for col in customers.columns:
    missing_percentage = customers[col].isnull().sum() / len(customers) * 100
    print(f"{col}: {missing_percentage:.2f}% missing values")
```

```python
threshold = 0.3  # Set a threshold (30% missing data)
rows_to_drop = customers[customers.isnull().mean(axis=1) > threshold]
customers_cleaned = customers.drop(rows_to_drop.index)


```

```python
customers_cleaned.isnull().sum()
```

```python
#fillna with mean
customers_cleaned['AVG_DAYS_BETWEEN_PURCHASES'] = customers_cleaned['AVG_DAYS_BETWEEN_PURCHASES'].fillna(customers_cleaned['AVG_DAYS_BETWEEN_PURCHASES'].mean())
customers_cleaned['MEDIAN_DAYS_BETWEEN_PURCHASES'] = customers_cleaned['MEDIAN_DAYS_BETWEEN_PURCHASES'].fillna(customers_cleaned['MEDIAN_DAYS_BETWEEN_PURCHASES'].mean())

# Fill NaN in 'first_purchase_day' with 'registration_date'
customers_cleaned['FIRST_PURCHASE_DAY'] = customers_cleaned['FIRST_PURCHASE_DAY'].fillna(customers_cleaned['REGISTRATION_DATE'])
# Fill NaN in 'last_purchase_day' with 'first_purchase_day'
customers_cleaned['LAST_PURCHASE_DAY'] = customers_cleaned['LAST_PURCHASE_DAY'].fillna(customers_cleaned['FIRST_PURCHASE_DAY'])
```

```python
customers_cleaned.isnull().sum()
```

```python
# Replace NaNs or empty lists in preferred restaurant type with 'unknown'
import ast

#Clean up unwanted characters (like \n, extra spaces, brackets)
def clean_restaurant_type(val):
    if pd.isna(val):
        return "unknown"
    try:
        # Fix string formatting and parse it as a list
        cleaned_val = ast.literal_eval(val.replace("\\n", "").replace("\\", "").strip())
        if isinstance(cleaned_val, list) and cleaned_val:
            return cleaned_val[0].strip().lower()  # return the first item in the list
        else:
            return "unknown"
    except Exception:
        return "unknown"

# Apply to column
customers_cleaned['PREFERRED_RESTAURANT_TYPES'] = customers_cleaned['PREFERRED_RESTAURANT_TYPES'].apply(clean_restaurant_type)
```

```python
customers_cleaned.head()
```

```python
#Changing datatype from Float to int for necessary columns
customers_cleaned['PURCHASE_COUNT_DELIVERY'] = (customers_cleaned['PURCHASE_COUNT_DELIVERY']).fillna(0).astype(int)
customers_cleaned['PURCHASE_COUNT_TAKEAWAY'] = (customers_cleaned['PURCHASE_COUNT_TAKEAWAY']).fillna(0).astype(int)
customers_cleaned['BREAKFAST_PURCHASES'] = (customers_cleaned['BREAKFAST_PURCHASES']).fillna(0).astype(int)
customers_cleaned['LUNCH_PURCHASES'] = (customers_cleaned['LUNCH_PURCHASES']).fillna(0).astype(int)
customers_cleaned['DINNER_PURCHASES'] = (customers_cleaned['DINNER_PURCHASES']).fillna(0).astype(int)
customers_cleaned['EVENING_PURCHASES'] = (customers_cleaned['EVENING_PURCHASES']).fillna(0).astype(int)
customers_cleaned['LATE_NIGHT_PURCHASES'] = (customers_cleaned['LATE_NIGHT_PURCHASES']).fillna(0).astype(int)
customers_cleaned['DISTINCT_PURCHASE_VENUE_COUNT'] = (customers_cleaned['DISTINCT_PURCHASE_VENUE_COUNT']).fillna(0).astype(int)
customers_cleaned['IOS_PURCHASES'] = (customers_cleaned['IOS_PURCHASES']).fillna(0).astype(int)
customers_cleaned['WEB_PURCHASES'] = (customers_cleaned['WEB_PURCHASES']).fillna(0).astype(int)
customers_cleaned['ANDROID_PURCHASES'] = (customers_cleaned['ANDROID_PURCHASES']).fillna(0).astype(int)
customers_cleaned['MOST_COMMON_HOUR_OF_THE_DAY_TO_PURCHASE'] = (customers_cleaned['MOST_COMMON_HOUR_OF_THE_DAY_TO_PURCHASE'].fillna(0)).astype(int)
customers_cleaned['MOST_COMMON_WEEKDAY_TO_PURCHASE'] = (customers_cleaned['MOST_COMMON_WEEKDAY_TO_PURCHASE']).fillna(0).astype(int)
customers_cleaned['AVG_DAYS_BETWEEN_PURCHASES'] = (customers_cleaned['AVG_DAYS_BETWEEN_PURCHASES']).fillna(0).astype(int)
customers_cleaned['MEDIAN_DAYS_BETWEEN_PURCHASES'] = (customers_cleaned['MEDIAN_DAYS_BETWEEN_PURCHASES']).fillna(0).astype(int)
```

```python
customers_cleaned.info()
```

```python
customers_cleaned['TOTAL_PURCHASES_EUR'] = customers_cleaned['TOTAL_PURCHASES_EUR'].fillna(customers_cleaned['TOTAL_PURCHASES_EUR'].mean())
customers_cleaned['MIN_PURCHASE_VALUE_EUR'] = customers_cleaned['MIN_PURCHASE_VALUE_EUR'].fillna(customers_cleaned['MIN_PURCHASE_VALUE_EUR'].mean())
customers_cleaned['MAX_PURCHASE_VALUE_EUR'] = customers_cleaned['MAX_PURCHASE_VALUE_EUR'].fillna(customers_cleaned['MAX_PURCHASE_VALUE_EUR'].mean())
customers_cleaned['AVG_PURCHASE_VALUE_EUR'] = customers_cleaned['AVG_PURCHASE_VALUE_EUR'].fillna(customers_cleaned['AVG_PURCHASE_VALUE_EUR'].mean())
customers_cleaned['AVERAGE_DELIVERY_DISTANCE_KMS'] = customers_cleaned['AVERAGE_DELIVERY_DISTANCE_KMS'].fillna(customers_cleaned['AVERAGE_DELIVERY_DISTANCE_KMS'].mean())
```

```python
# Segment users based on AVG_DAYS_BETWEEN_PURCHASES
customers_cleaned['Purchase_Frequency'] = pd.cut(
    customers_cleaned['AVG_DAYS_BETWEEN_PURCHASES'],
    bins=[0, 7, 30, float('inf')],
    labels=['Frequent', 'Moderate', 'Infrequent']
)

# Compare segments with TOTAL_PURCHASES_EUR
segment_spending = customers_cleaned.groupby('Purchase_Frequency')['TOTAL_PURCHASES_EUR'].mean()

# Print the average spending for each segment
print(segment_spending)

# Visualize the spending behavior across segments
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x=segment_spending.index, y=segment_spending.values, palette=['#4C4A59', '#1B7F7A', '#0897B4'])
plt.title('Average Spending by Purchase Frequency Segment', fontsize=16)
plt.xlabel('Purchase Frequency', fontsize=12)
plt.ylabel('Average Total Spending (€)', fontsize=12)

# Add legend
handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in ['#4C4A59', '#1B7F7A', '#0897B4']]
labels = ['Frequent', 'Moderate', 'Infrequent']
plt.legend(handles, labels, title="Segments", loc='upper right')

plt.show()

# Purchase Frequency Segmentation
total_revenue_by_segment = customers_cleaned.groupby('Purchase_Frequency')['TOTAL_PURCHASES_EUR'].sum()
print(total_revenue_by_segment)
```
